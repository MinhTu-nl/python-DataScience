## 1. Problem Definition
- Goal: predict the sales price for each house

## 2. Feature Selection:
- Choose feature to train ML Model

## 3. Spliting the dataset
 3.1: dataset -> x,y
- `data` : dataset
- 'x' : data[features]

 3.2 X,y -> X_train, y_train, X_vaild, y_vaild

 ## 4. Training Machine Learning Model

```python
import pandas as pd
import numpy as np
```

data = pd.read_csv('train.csv', index_col = 'Id')
data.head()

```python
data.columns
```

### 2. Feature Selection:

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
```

### 3. Spliting dataset into X and y

```python
X = data[features]
y = data['SalePrice']
X.head()
```

```python
y.head()
```

### 3.2 X,y -> X_train, y_train, X_vaild, y_vaild

```python
from sklearn.model_selection import train_test_split
X_train, X_vaild, y_train,y_vaild = train_test_split(X, y, train_size = 0.8, test_size = 0.2, random_state =0 )
```

```python
X_train
```

```python
X_train.shape
```

```python
data.shape
```

### 4. Training Machine Learning Model

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)
```

```python
#fit training data into model
dt_model.fit(X_train, y_train)
```

```python
y_preds = dt_model.predict(X_vaild.head())
y_preds
```

```python
pd.DataFrame({'y' : y_vaild.head(), 'y_preds' : y_preds})
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
rf_model = RandomForestRegressor(random_state = 1)
rf_model.fit(X_train, y_train)
```

```python
rf_vaild_preds = rf_model.predict(X_vaild)
```
