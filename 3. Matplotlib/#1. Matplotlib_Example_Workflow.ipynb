## Matplotlib Example Workflow 
Quy trình làm việc ví dụ Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#1. Prepare data - chuẩn bị dữ liệu
x = [1,2,3,4]
y = [11,22,33,44]
```

```python
#2 setup plot - sơ đồ thiết lập
fig, ax = plt.subplots(figsize=(5,5)) 
# Kích thước hình = Chiều rộng & Chiều cao của ô

#3 Plot data - vẽ dữ liệu
ax.plot(x,y);

#4 Customize plot - TUỲ CHỈNH HÌNH VẼ
ax.set(title = 'Sample simple plot',
                xlabel = 'x',
                ylabel = 'y')

# 5. Save & show
#fig.savefig("../images/simple-plot.png")
```
